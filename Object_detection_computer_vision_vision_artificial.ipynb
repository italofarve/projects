Example of Real-Time Object Detection using Yolo

Yolo is a real-time object detection algorithm that identifies specific objects in videos, live feeds, or images. YOLO uses features learned by a deep convolutional neural network to detect an object.

The model was trained using Google Cloud (Google Colab), then I use the model for run in a Macbook to detecting object in a video of New York City, as a: cars, traffic ligths and persons.

The model also works using the camera of Macbook and detected in real-time persons a bottles.
